1. What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a formal procedure used to make decisions or inferences about a population parameter based on sample data. It involves setting up two competing hypotheses: a null hypothesis (H0) and an alternative hypothesis (H1). The null hypothesis typically represents a statement of no effect or no difference, while the alternative hypothesis represents what we are trying to find evidence for.



2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
- The null hypothesis (H0) and the alternative hypothesis (H1 or Ha) are two competing statements used in hypothesis testing.

- Null Hypothesis (H0): This is a statement of no effect, no difference, or no relationship between variables. It represents the status quo or a commonly accepted belief. Researchers typically aim to gather evidence against the null hypothesis. For example, if you're testing a new drug, the null hypothesis might be that the drug has no effect on a patient's condition.

- Alternative Hypothesis (H1 or Ha): This is a statement that contradicts the null hypothesis. It represents what the researcher is trying to find evidence for or what they believe might be true. It often suggests an effect, a difference, or a relationship. Using the drug example, the alternative hypothesis might be that the drug does have a positive effect on the patient's condition.

- Key Differences:

1. Purpose: H0 is the statement to be tested and potentially rejected, while H1 is the statement you are trying to support with evidence.
2. Assumption: In hypothesis testing, we initially assume the null hypothesis is true until there is sufficient evidence to reject it.
3. Outcome: If you reject H0, it means you have statistical evidence to support H1. If you fail to reject H0, it means you don't have enough statistical evidence to support H1 (it does not mean H0 is true).
4. Mathematical Form: H0 typically includes an equality sign (e.g., =, ≤, ≥), while H1 typically includes an inequality sign (e.g., ≠, <, >).

3. : Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
- The significance level, often denoted by the Greek letter alpha (α), is a crucial concept in hypothesis testing. It represents the probability of rejecting the null hypothesis (H0) when it is actually true. In simpler terms, it's the maximum acceptable probability of making a Type I error.
- The significance level (alpha, α) is the probability of rejecting a true null hypothesis (making a Type I error). It's typically set at 0.05 or 0.01.

Its role is to provide a threshold for deciding the test outcome:

- If the p-value (probability of observing your data if the null hypothesis were true) is less than or equal to alpha, you reject the null hypothesis.
- If the p-value is greater than alpha, you fail to reject the null hypothesis.

4. What are Type I and Type II errors? Give examples of each.

I. Type I Error (False Positive)
- Definition: A Type I error occurs when you reject a true null hypothesis (H0). It's like concluding there is an effect or difference when, in reality, there isn't one.
- Probability: The probability of making a Type I error is denoted by α (alpha), which is the significance level you set for your test (e.g., 0.05 or 5%).
- Analogy: In a legal trial, a Type I error would be convicting an innocent person.
- Example: A medical test claims a patient has a disease (reject H0: patient is healthy) when in fact, the patient is healthy (H0 is true).

II. Type II Error (False Negative)
- Definition: A Type II error occurs when you fail to reject a false null hypothesis (H0). It's like concluding there is no effect or difference when, in reality, there is one.
- Probability: The probability of making a Type II error is denoted by β (beta). The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.
- Analogy: In a legal trial, a Type II error would be letting a guilty person go free.
- Example: A medical test claims a patient is healthy (fail to reject H0: patient is healthy) when in fact, the patient has the disease (H0 is false).

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
- The Z-test and T-test are both parametric statistical tests used to compare means. The main difference between them lies in the information you have about the population standard deviation and the sample size.

I. Z-test :

- When to Use: A Z-test is typically used when:

1. The population standard deviation (σ) is known.
2. The sample size (n) is large (generally n ≥ 30), even if the population standard deviation is unknown (due to the Central Limit Theorem, the sample standard deviation approximates the population standard deviation for large samples, and the sampling distribution of the mean approaches a normal distribution).
- Assumptions: The data is normally distributed, or the sample size is large enough for the Central Limit Theorem to apply.

- Formula (for a single population mean): Z = (x̄ - μ) / (σ / √n)

x̄: Sample mean
μ: Population mean (hypothesized)
σ: Population standard deviation
n: Sample size
- Example: You want to test if the average height of adult men in a city is 175 cm. You know from historical data that the population standard deviation of heights is 7 cm. You take a random sample of 100 men.

II. T-test :

- When to Use: A T-test is used when:

1. The population standard deviation (σ) is unknown.
2. The sample size (n) is small (generally n < 30). For larger samples, even if σ is unknown, a Z-test can often be applied, but a T-test is technically more appropriate when σ is unknown, regardless of sample size, as it accounts for the additional uncertainty.
- Assumptions: The data is approximately normally distributed (especially critical for small sample sizes). The t-distribution has heavier tails than the normal distribution, which accounts for the greater uncertainty with smaller samples and unknown population standard deviation.

- Formula (for a single population mean): t = (x̄ - μ) / (s / √n)

x̄: Sample mean
μ: Population mean (hypothesized)
s: Sample standard deviation
n: Sample size
- Example: You want to test if a new teaching method improves test scores. You don't know the population standard deviation of test scores for this method. You try the new method on a small group of 25 students and record their scores.

### Binomial Distribution Generation and Histogram Plotting

This code generates random samples from a binomial distribution and visualizes their frequency using a histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success on each trial
size = 10000 # Number of times to repeat the experiment (number of samples)

# Generate random samples from a binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(binomial_data, bins=np.arange(n + 2) - 0.5, edgecolor='black', density=True)
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(np.arange(n + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Z-Test Implementation for Hypothesis Testing

This code performs a one-sample Z-test to determine if the mean of the `sample_data` is significantly different from a hypothesized population mean. For a Z-test, we assume the population standard deviation is known.

```python
import numpy as np
from scipy import stats

# Sample data provided by the user
sample_data = np.array([
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
])

# --- Hypothesis Formulation ---
# Null Hypothesis (H0): The true population mean (μ) is equal to 50.
# H0: μ = 50

# Alternative Hypothesis (H1): The true population mean (μ) is not equal to 50.
# H1: μ ≠ 50 (Two-tailed test)

hypothesized_population_mean = 50

# --- Parameters for Z-test ---
# For a Z-test, the population standard deviation (sigma) is assumed to be known.
# We will assume a value for demonstration purposes.
population_std_dev = 1.0 # Assumed known population standard deviation
significance_level = 0.05 # Alpha (α)

# --- Sample Statistics ---
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Size: {sample_size}")
print(f"Hypothesized Population Mean (μ0): {hypothesized_population_mean}")
print(f"Assumed Population Standard Deviation (σ): {population_std_dev}")
print(f"Significance Level (α): {significance_level}")
print("\n--- Performing Z-Test ---")

# Calculate the Z-statistic
# Z = (sample_mean - hypothesized_population_mean) / (population_std_dev / sqrt(sample_size))
z_statistic = (sample_mean - hypothesized_population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
# For a two-tailed test, we multiply by 2 because extreme values can be in either tail
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

print(f"Z-Statistic: {z_statistic:.3f}")
print(f"P-Value: {p_value:.3f}")

# --- Decision ---
print("\n--- Decision ---")
if p_value < significance_level:
    print(f"Since the P-Value ({p_value:.3f}) is less than the Significance Level ({significance_level}), we Reject the Null Hypothesis.")
    print("Conclusion: There is sufficient evidence to conclude that the sample mean is significantly different from the hypothesized population mean of {hypothesized_population_mean}.")
else:
    print(f"Since the P-Value ({p_value:.3f}) is greater than or equal to the Significance Level ({significance_level}), we Fail to Reject the Null Hypothesis.")
    print("Conclusion: There is not sufficient evidence to conclude that the sample mean is significantly different from the hypothesized population mean of {hypothesized_population_mean}.")
```

### Normal Distribution Simulation and Confidence Interval Calculation

This script generates data from a normal distribution, calculates the 95% confidence interval for its mean, and visualizes the data with the confidence interval on a histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --- Parameters for Data Simulation ---
population_mean = 100       # True mean of the underlying normal distribution
population_std_dev = 15     # True standard deviation of the underlying normal distribution
sample_size = 1000          # Number of data points to simulate
confidence_level = 0.95     # Desired confidence level for the interval (e.g., 95%)

# --- Simulate Data from a Normal Distribution ---
np.random.seed(42) # for reproducibility
simulated_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

print(f"Simulated Data Mean (approximate): {np.mean(simulated_data):.2f}")
print(f"Simulated Data Std Dev (approximate): {np.std(simulated_data):.2f}")

# --- Calculate Sample Statistics ---
sample_mean = np.mean(simulated_data)
sample_std_dev = np.std(simulated_data, ddof=1) # Use ddof=1 for sample standard deviation

# --- Calculate 95% Confidence Interval for the Mean ---
# For confidence intervals of the mean when population std dev is unknown (common case),
# we use the t-distribution.

# Degrees of freedom
df = sample_size - 1

# Calculate the standard error of the mean (SEM)
sem = sample_std_dev / np.sqrt(sample_size)

# Calculate the confidence interval
confidence_interval = stats.t.interval(confidence_level, df, loc=sample_mean, scale=sem)

print(f"\nSample Mean: {sample_mean:.3f}")
print(f"Sample Standard Deviation: {sample_std_dev:.3f}")
print(f"Standard Error of the Mean (SEM): {sem:.3f}")
print(f"{int(confidence_level*100)}% Confidence Interval for the Mean: ({confidence_interval[0]:.3f}, {confidence_interval[1]:.3f})")

# --- Plot the Data ---
plt.figure(figsize=(12, 7))

# Histogram of the simulated data
plt.hist(simulated_data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Simulated Data Histogram')

# Plot the sample mean
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.3f}')

# Plot the confidence interval
plt.axvline(confidence_interval[0], color='blue', linestyle='dotted', linewidth=2, label=f'{int(confidence_level*100)}% CI Lower Bound: {confidence_interval[0]:.3f}')
plt.axvline(confidence_interval[1], color='blue', linestyle='dotted', linewidth=2, label=f'{int(confidence_level*100)}% CI Upper Bound: {confidence_interval[1]:.3f}')

# Add shaded region for confidence interval
plt.axvspan(confidence_interval[0], confidence_interval[1], color='blue', alpha=0.1, label=f'{int(confidence_level*100)}% CI Region')

plt.title(f'Simulated Normal Distribution Data with {int(confidence_level*100)}% Confidence Interval (n={sample_size})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Z-Score Calculation and Visualization

This script defines a function to calculate Z-scores for a given dataset and then visualizes the distribution of these Z-scores using a histogram. It also includes an explanation of what Z-scores signify.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculates the Z-scores for each data point in a given dataset.

    Parameters:
    data (array-like): The input dataset.

    Returns:
    numpy.ndarray: An array of Z-scores.
    """
    data_array = np.asarray(data)
    mean = np.mean(data_array)
    std_dev = np.std(data_array) # Population standard deviation for Z-score calculation

    # Avoid division by zero if standard deviation is 0 (e.g., all data points are the same)
    if std_dev == 0:
        return np.zeros_like(data_array)

    z_scores = (data_array - mean) / std_dev
    return z_scores

# --- Example Usage ---
# Using the simulated_data from previous steps (if available in the environment)
# Otherwise, create a new sample dataset
if 'simulated_data' in locals():
    dataset = simulated_data
    data_label = "Simulated Normal Data"
else:
    np.random.seed(42) # for reproducibility
    dataset = np.random.normal(loc=50, scale=10, size=1000)
    data_label = "Randomly Generated Normal Data"

print(f"Original {data_label} Mean: {np.mean(dataset):.2f}")
print(f"Original {data_label} Standard Deviation: {np.std(dataset):.2f}")

z_scores = calculate_z_scores(dataset)

print(f"\nZ-Scores Mean: {np.mean(z_scores):.2f} (should be close to 0)")
print(f"Z-Scores Standard Deviation: {np.std(z_scores):.2f} (should be close to 1)")

# --- Visualize the standardized data ---
plt.figure(figsize=(10, 6))
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title(f'Histogram of Z-Scores for {data_label}')
plt.xlabel('Z-Score')
plt.ylabel('Density')
plt.axvline(0, color='red', linestyle='--', label='Mean (0)')
plt.axvline(1, color='green', linestyle=':', label='1 Std Dev')
plt.axvline(-1, color='green', linestyle=':', label='-1 Std Dev')
plt.axvline(2, color='orange', linestyle=':', label='2 Std Dev')
plt.axvline(-2, color='orange', linestyle=':', label='-2 Std Dev')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
